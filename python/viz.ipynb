```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for Rust and C++ (Natasja's Computer)
rust_allocation_natasja = [0.813893, 0.712036, 0.902309, 0.777595, 0.674618, 0.590551, 0.567738, 0.690611, 0.7854, 0.640146]
rust_deallocation_natasja = [0.955077, 0.844493, 0.985369, 0.851275, 0.799496, 0.761724, 0.806435, 0.844414, 0.776423, 0.828466]

# Data for C++ (Natasja's Computer)
cpp_allocation_natasja = [2.2055, 2.18095, 2.11933, 1.72855, 1.74165, 1.40541, 1.36401, 1.32834, 0.946323, 1.1954]
cpp_deallocation_natasja = [5.17301, 2.18376, 3.07298, 1.91375, 1.85872, 1.85118, 1.84611, 1.7746, 1.85787, 1.83771]

# Data for Rust and C++ (Lasse's Computer)
rust_allocation_lasse = [0.891852, 0.799833, 0.79712, 0.802402, 0.802465, 0.81924, 0.799957, 0.811108, 0.805714, 0.81188]
rust_deallocation_lasse = [0.918833, 0.847299, 0.831308, 0.850416, 0.869381, 0.855372, 0.878636, 0.843761, 0.87174, 0.862089]

# Data for C++ (Lasse's Computer)
cpp_allocation_lasse = [0.819672, 0.808291, 0.854868, 0.814293, 0.820071, 0.808066, 0.810093, 0.815091, 0.810827, 0.814675]
cpp_deallocation_lasse = [1.23806, 1.21419, 1.20707, 1.19816, 1.21932, 1.16079, 1.19942, 1.2144, 1.18579, 1.22263]

# Test case numbers (1 to 10)
test_cases = list(range(1, 11))

# Create DataFrame to make plotting easier (for both Natasja and Lasse)
data_natasja = pd.DataFrame({
    'Test Case': test_cases,
    'Rust Allocation Time': rust_allocation_natasja,
    'Rust Deallocation Time': rust_deallocation_natasja,
    'C++ Allocation Time': cpp_allocation_natasja,
    'C++ Deallocation Time': cpp_deallocation_natasja
})

data_lasse = pd.DataFrame({
    'Test Case': test_cases,
    'Rust Allocation Time': rust_allocation_lasse,
    'Rust Deallocation Time': rust_deallocation_lasse,
    'C++ Allocation Time': cpp_allocation_lasse,
    'C++ Deallocation Time': cpp_deallocation_lasse
})

# Plot Allocation and Deallocation Time for Natasja's Computer
plt.figure(figsize=(10, 6))
plt.plot(data_natasja['Test Case'], data_natasja['Rust Allocation Time'], label="Rust Allocation Time", marker='o', color='b')  # Rust = blue
plt.plot(data_natasja['Test Case'], data_natasja['Rust Deallocation Time'], label="Rust Deallocation Time", marker='o', linestyle='--', color='b')  # Rust Deallocation = blue dashed
plt.plot(data_natasja['Test Case'], data_natasja['C++ Allocation Time'], label="C++ Allocation Time", marker='o', color='orange')  # C++ = orange
plt.plot(data_natasja['Test Case'], data_natasja['C++ Deallocation Time'], label="C++ Deallocation Time", marker='o', linestyle='--', color='orange')  # C++ Deallocation = orange dashed

plt.title('Natasja - Allocation and Deallocation Time Comparison (Rust vs C++)')
plt.xlabel('Test Case')
plt.ylabel('Time (s)')
plt.legend()
plt.grid(True)
plt.show()

# Plot Allocation and Deallocation Time for Lasse's Computer
plt.figure(figsize=(10, 6))
plt.plot(data_lasse['Test Case'], data_lasse['Rust Allocation Time'], label="Rust Allocation Time", marker='o', color='b')  # Rust = blue
plt.plot(data_lasse['Test Case'], data_lasse['Rust Deallocation Time'], label="Rust Deallocation Time", marker='o', linestyle='--', color='b')  # Rust Deallocation = blue dashed
plt.plot(data_lasse['Test Case'], data_lasse['C++ Allocation Time'], label="C++ Allocation Time", marker='o', color='orange')  # C++ = orange
plt.plot(data_lasse['Test Case'], data_lasse['C++ Deallocation Time'], label="C++ Deallocation Time", marker='o', linestyle='--', color='orange')  # C++ Deallocation = orange dashed

plt.title('Lasse - Allocation and Deallocation Time Comparison (Rust vs C++)')
plt.xlabel('Test Case')
plt.ylabel('Time (s)')
plt.legend()
plt.grid(True)
plt.show()
```
